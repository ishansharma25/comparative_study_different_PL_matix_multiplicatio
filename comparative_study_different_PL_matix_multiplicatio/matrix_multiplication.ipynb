```python
import numpy as np
import time

def multiply_matrices(matrix1, matrix2):
    return np.dot(matrix1, matrix2)



n= int(input("Enter choice: "))
if(n==1):
    size1 = (1000, 1000)
    size2 = (1000, 1000)
elif(n==2):
    size1 = (2000, 2000)
    size2 = (2000, 2000)
elif(n==3):
    size1 = (2500, 2500)
    size2 = (2500, 2500)
elif(n==4):
    size1 = (3000, 3000)
    size2 = (3000, 3000)
elif(n==5):
    size1 = (3500, 3500)
    size2 = (3500, 3500)





```

```python

matrix1 = np.random.rand(size1[0], size1[1])
matrix2 = np.random.rand(size2[0], size2[1])


print(len(matrix1)*len(matrix1[0]))
print(len(matrix2)*len(matrix2[0]))
```

```python

start_time = time.time()
result_matrix = multiply_matrices(matrix1, matrix2)
end_time = time.time()


execution_time = end_time - start_time

print("Execution time:", execution_time, "seconds")
```
